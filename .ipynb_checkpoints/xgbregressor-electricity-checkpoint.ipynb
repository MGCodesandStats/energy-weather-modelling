```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pmdarima as pm
from sklearn.metrics import mean_squared_error
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.tsa.stattools as ts

# fix random seed for reproducibility
np.random.seed(7)

# load dataset
d1 = pd.read_csv('dccelectricitycivicsblocks34p20130221-1840.csv', engine='python', skipfooter=3)
d1
```

```python
d2=d1.rename(columns=d1.iloc[0])
d3=d2.drop(d1.index[0])
d3
d3.drop(d3.index[0])
d4=d3.drop('Date', axis=1)
d5=d4.drop('Values', axis=1)
d5
d6=d5.dropna()
d7=d6.values
d7
df=np.sum(d7, axis=1, dtype=float)
df
```

```python
df=np.array(df)
df
```

```python
def create_dataset(df, previous=1):
    dataX, dataY = [], []
    for i in range(len(df)-previous-1):
        a = df[i:(i+previous), 0]
        dataX.append(a)
        dataY.append(df[i + previous, 0])
    return np.array(dataX), np.array(dataY)
```

```python
df=df.reshape(-1,1)
df
```

```python
train_size = int(len(df) * 0.8)
train_size
```

```python
test_size = len(df) - train_size
test_size
```

```python
train, test = df[0:train_size,:], df[train_size:len(df),:]
```

```python
train
```

```python
test
```

```python
train.shape
```

```python
test.shape
```

```python
# Lookback period
lookback = 7
X_train, Y_train = create_dataset(train, lookback)
X_test, Y_test = create_dataset(test, lookback)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
model.fit(X_train, Y_train)
```

```python
testpred = model.predict(X_test)
```

```python
Y_test=Y_test.reshape(-1,1)
testpred=testpred.reshape(-1,1)
```

```python
import math
from math import sqrt
test_mse = mean_squared_error(Y_test, testpred)
rmse = sqrt(test_mse)
print('RMSE: %f' % rmse)
```
