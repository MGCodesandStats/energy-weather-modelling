### Import Libraries and define dtypes

```python
import csv
import math
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import pmdarima as pm
import random
import statsmodels.api as sm
import statsmodels.formula.api as smf

from numpy.random import seed
seed(1)
from pandas import DataFrame
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.tsa.stattools as ts
```

```python
df = pd.read_csv('meteireann_dublin.csv')
a=df.head()
a
```

```python
tseries=df['meant']
tseries
```

```python
plt.plot(tseries)
plt.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False,
    labelbottom=False)
plt.show()
```

### Train-Test Split

```python
tseriesr=pd.Series(tseries)
type(tseriesr)
train, test = tseriesr[1:732], tseriesr[733:915]
```

```python
df=pd.DataFrame(tseries)
df
```

### Autocorrelation and Partial Autocorrelation Plots

```python
acf(tseries, nlags=30)
```

```python
plot_acf(tseries, lags=30, zero=False);
```

```python
pacf(tseries, nlags=100)
```

```python
plot_pacf(tseries, lags=10, zero=False);
```

### ARIMA

```python
model=sm.tsa.statespace.SARIMAX(endog=train,order=(1,0,0),seasonal_order=(2,1,0,12),trend='c',enforce_invertibility=False)
results=model.fit()
print(results.summary())
```

```python
len(test)
```

```python
prediction=pd.DataFrame(results.predict(n_periods=183), index=test.index)
prediction
```

```python
prediction.columns = ['Predicted_Temperature']
predictions=prediction['Predicted_Temperature']
```

```python
test
```

```python
plt.figure(figsize=(15,10))
plt.plot(train, label='Training')
plt.plot(test, label='Validation')
plt.plot(prediction, label='Predicted')
plt.legend(loc = 'upper center')
plt.show()
```

### Calculate Test RMSE

```python
mse = mean_squared_error(test, predictions)
rmse = math.sqrt(mse)
print('RMSE: %f' % rmse)
```
